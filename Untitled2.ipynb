```python
import dashscope
from dashscope import TextEmbedding, Generation
import numpy as np
import os
dashscope.api_key = "sk-9b7fdff3e2674d34ae8bf409d39205e7"

# 准备知识文档库（浙江交通职业技术学院与AI相关知识）
knowledge_base = [
    # 学院基本信息（1-10）
    "浙江交通职业技术学院创建于1958年，是交通运输部与浙江省人民政府共建的高职院校。",
    "浙江交通职业技术学院位于杭州市莫干山路1515号，地理位置优越。",
    "浙江交通职业技术学院的校训是'励志力行'，强调实践与奋斗精神。",
    "学院是国家骨干高职院校建设单位，浙江省示范性高等职业院校。",
    "学院设有路桥学院、汽车学院、海运学院、航空学院等多个二级学院。",
    "浙江交通职业技术学院拥有国家级教学团队和省级重点实验室。",
    "学院注重产教融合，与多家知名企业建立了深度合作关系。",
    "学院在智慧交通领域处于省内领先地位，培养了大量技术技能人才。",
    "浙江交通职业技术学院的毕业生就业率连续多年保持在98%以上。",
    "学院积极推进国际化办学，与多个国家的院校建立合作关系。",

    # 专业设置（11-20）
    "道路桥梁工程技术是学院的王牌专业，培养高素质的交通建设人才。",
    "汽车检测与维修技术专业配备了先进的实训设备和专业师资。",
    "智能交通技术专业紧跟时代发展，培养智慧交通系统管理人才。",
    "航海技术专业是浙江省特色专业，培养现代航运管理人才。",
    "飞机机电设备维修专业与多家航空公司建立了订单式培养模式。",
    "城市轨道交通运营管理专业适应杭州地铁快速发展的需求。",
    "新能源汽车技术专业顺应汽车产业电动化转型趋势。",
    "物流管理专业培养现代供应链管理的专业人才。",
    "人工智能技术应用专业是学院新开设的热门专业。",
    "大数据技术专业培养数据分析和处理的技术人才。",

    # AI基础知识（21-35）
    "人工智能是计算机科学的一个分支，致力于创建能够模拟人类智能的系统。",
    "机器学习是AI的核心技术，让计算机从数据中自动学习规律。",
    "深度学习是机器学习的子领域，使用多层神经网络来学习数据特征。",
    "神经网络模仿人脑神经元的连接方式，用于模式识别和决策。",
    "自然语言处理让计算机能够理解和生成人类语言。",
    "计算机视觉使机器能够'看懂'图像和视频内容。",
    "强化学习通过奖励和惩罚机制让AI学会最优策略。",
    "迁移学习可以将一个领域学到的知识应用到另一个领域。",
    "生成式AI能够创造全新的内容，如文本、图像、音乐等。",
    "大语言模型通过学习海量文本数据来理解和生成语言。",
    "Transformer架构是现代NLP模型的基础，具有强大的序列处理能力。",
    "注意力机制让模型能够关注输入中最重要的部分。",
    "预训练模型通过大规模数据预先训练，可以fine-tune用于特定任务。",
    "提示工程是引导AI模型生成期望输出的技术。",
    "向量数据库专门用于存储和检索高维向量数据。",

    # RAG和应用技术（36-45）
    "RAG技术可以让AI模型在回答问题时参考外部知识库，提高回答的准确性。",
    "向量化是将文本、图像等数据转换为数值向量的过程。",
    "嵌入向量能够捕捉语义信息，相似的内容会有相似的向量表示。",
    "余弦相似度是衡量两个向量相似程度的常用方法。",
    "检索增强生成结合了信息检索和文本生成的优势。",
    "知识图谱将知识结构化表示为实体和关系的网络。",
    "智能问答系统能够理解用户问题并给出准确答案。",
    "对话系统可以与用户进行多轮自然对话交互。",
    "AI Agent是能够自主执行任务的智能代理。",
    "多模态AI可以同时处理文本、图像、音频等多种类型的数据。",

    # 交通领域AI应用（46-50）
    "智能交通系统利用AI技术优化交通流量，减少拥堵。",
    "自动驾驶技术综合运用计算机视觉、深度学习等AI技术。",
    "智慧物流通过AI优化配送路线，提高运输效率。",
    "交通流量预测使用机器学习模型预测未来交通状况。",
    "智能信号灯控制系统根据实时交通流量动态调整信号配时。"
    ]
print("📚 知识库准备完成！")
print(f"共有 {len(knowledge_base)} 条知识")
print("\n📑 知识分类：")
print("  • 学院信息：10条")
print("  • 专业设置：10条")
print("  • AI基础：15条")
print("  • RAG技术：10条")
print("  • 交通应用：5条")
print("\n显示前5条示例：")
for i in range(5):
    print(f"  {i+1}. {knowledge_base[i][:30]}...")

def text_to_vector(text):

    try:
        response = TextEmbedding.call(
            model='text-embedding-v2',  # 使用阿里云的嵌入模型
            input=text
        )
        return response.output['embeddings'][0]['embedding']
    except Exception as e:
        print(f"❌ 错误：{e}")
        print("请检查API密钥是否正确设置")
        return None
# 将所有知识文档转换为向量
print("🔄 正在将知识库转换为向量...")
print("提示：50条知识需要一些时间，请耐心等待\n")
knowledge_vectors = []
for i, doc in enumerate(knowledge_base, 1):
    vector = text_to_vector(doc)
    if vector:
        knowledge_vectors.append(vector)
        # 每10条显示一次进度
        if i % 10 == 0:
            print(f"✅ 已完成 {i}/50 条")
            print(f"\n🎯 转换完成！")
            print(f"• 共转换 {len(knowledge_vectors)} 条知识")
            print(f"• 每条知识被转换为 {len(knowledge_vectors[0]) if knowledge_vectors else 0} 维向量")

def calculate_similarity(vec1, vec2):
    """计算两个向量的余弦相似度"""
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    # 余弦相似度 = 向量点积 / (向量模长的乘积)
    cosine = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
    return cosine
print("📐 相似度计算函数准备完成！")

def retrieve_relevant_docs(query, top_k=3):
    """根据查询检索最相关的文档"""
    print(f"\n🔍 正在搜索与问题相关的知识...")
    # 1. 将查询转换为向量
    query_vector = text_to_vector(query)
    if not query_vector:
        return []

    # 2. 计算查询与所有文档的相似度
    similarities = []
    for i, doc_vector in enumerate(knowledge_vectors):
        sim = calculate_similarity(query_vector, doc_vector)
        similarities.append((i, sim, knowledge_base[i]))


    similarities.sort(key=lambda x: x[1], reverse=True)

# 4. 返回最相关的文档
    relevant_docs = []
    print(f"\n📊 相似度排名（前{top_k}个）：")
    for i in range(min(top_k, len(similarities))):
        idx, sim, doc = similarities[i]
        print(f"  {i+1}. 相似度: {sim:.3f} | {doc[:40]}...")
        relevant_docs.append(doc)

    return relevant_docs
print("✅ 检索功能准备完成！")

def rag_generate(query):
    """RAG生成：检索 + 生成"""
    print("="*60)
    print(f"❓ 用户问题：{query}")
    print("="*60)

    # 步骤1：检索相关文档
    relevant_docs = retrieve_relevant_docs(query, top_k=3)

    if not relevant_docs:
        return "抱歉，无法检索到相关信息。"

    # 步骤2：构建增强的提示词
    context = "\n".join([f"- {doc}" for doc in relevant_docs])

    prompt = f"""
    请基于以下参考信息回答问题。
    参考信息：{context} 
    问题：{query}
    请给出准确、简洁的回答：
    """

    print("\n💭 正在生成回答...")

    # 步骤3：调用大模型生成答案
    try:
        response = Generation.call(
            model='qwen-turbo',  # 使用通义千问模型
            prompt=prompt
        )
        answer = response.output.text
        print("\n✨ RAG生成的答案：")
        print("-"*60)
        print(answer)
        print("-"*60)
        return answer
    except Exception as e:
        return f"生成答案时出错：{e}"
print("🚀 RAG系统准备就绪！")
```
